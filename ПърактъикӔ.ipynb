```python
import PyPDF2
import pandas as pd
import re
import os

import nltk
nltk.download('punkt')
import re
from nltk.tokenize import word_tokenize, sent_tokenize
```

# **Сначала создадим датафрейм с газетами**

```python
#Функция для извлечения текста из газет pdf:
def extract_text_from_pdf(file_path):
    with open(file_path, 'rb') as file:
        pdf_reader = PyPDF2.PdfReader(file)
        num_pages = len(pdf_reader.pages)

        text = ''
        for page_number in range(num_pages):
            page = pdf_reader.pages[page_number]
            text += page.extract_text()

        return text
```

```python
#Функции для подсчета слов, уникальных слов и предложений
def count_words(text):
    words = word_tokenize(text)
    return len(words)

def count_unique_words(text):
    words = word_tokenize(text)
    unique_words = set(words)
    return len(unique_words)

def count_sentences(text):
    sentences = sent_tokenize(text)
    return len(sentences)
```

```python
# Создание пустого списка для хранения данных
data = []
```

```python
# Путь к папке с PDF-файлами газет
folder_path = '/content/drive/MyDrive/Практика_3_курс/Газеты_PDF'
```

```python
# Цикл для обхода файлов в папке
for filename in os.listdir(folder_path):
    if filename.endswith('.pdf'):
        file_path = os.path.join(folder_path, filename)

        # Извлечение текста из PDF-файла
        text = extract_text_from_pdf(file_path)

        # Получение количества слов, уникальных слов и предложений
        word_count = count_words(text)
        unique_word_count = count_unique_words(text)
        sentence_count = count_sentences(text)

        # Создание словаря с данными
        data.append({
            'Номер текста': filename,
            'Количество слов': word_count,
            'Количество уникальных слов': unique_word_count,
            'Количество предложений': sentence_count
        })
```

```python
# Создание датафрейма из списка данных
df = pd.DataFrame(data)
df
```

# **Теперь сделаем то же самое для файлов txt**

```python
# Путь к папке с очищенными текстовыми файлами
folder_path2 = "/content/drive/MyDrive/Практика_3_курс/Очищенный_текст "
```

```python
# Список для хранения данных о текстах
texts_data = []
```

```python
file_list = os.listdir(folder_path2)
```

```python
# Проходимся по каждому элементу
for file_name in file_list:
    # Формируем полный путь к элементу
    file_path2 = os.path.join(folder_path2, file_name)


    # Читаем содержимое файла с указанием правильной кодировки
    with open(file_path2, 'r', encoding='utf-8') as file:
        text = file.read()

    # Получаем информацию о тексте
    number_match = re.search(r'\d+', file_name)
    number = number_match.group() if number_match else ''  # Номер текста

    words = word_tokenize(text)
    word_count = len(words)  # Количество слов
    unique_words = len(set(words))  # Количество уникальных слов

    sentences = sent_tokenize(text)
    sentence_count = len(sentences)  # Количество предложений

    # Добавляем данные в список
    texts_data.append([number, word_count, unique_words, sentence_count])
```

```python
# Создаем датафрейм из списка данных
df2 = pd.DataFrame(texts_data, columns=['Номер текста', 'Количество слов', 'Количество уникальных слов', 'Количество предложений'])
df2
```

# **Готовый Датафрейм:**

```python
# Объединяем два датафрейма
GotovayaDataFrame = pd.concat([df, df2], ignore_index=True)
GotovayaDataFrame
```

```python
# Вычисляем сумму значений в столбце 'Количество слов'
slova = GotovayaDataFrame['Количество слов'].sum()
sentences = GotovayaDataFrame['Количество предложений'].sum()
uniq = GotovayaDataFrame['Количество уникальных слов'].sum()

print(slova, sentences, uniq)
```

```python
def save_dataframe_to_csv(dataframe, folder_path, file_name):
    # Создаем путь к файлу
    file_path = os.path.join(folder_path, file_name)

    # Сохраняем датафрейм в файл CSV
    dataframe.to_csv(file_path, index=False)


folder_path = '/content/drive/MyDrive/Практика_3_курс'
file_name = 'textdata2.csv'

save_dataframe_to_csv(GotovayaDataFrame, folder_path, file_name)
```
